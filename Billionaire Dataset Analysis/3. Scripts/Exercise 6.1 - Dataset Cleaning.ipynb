### Data wrangling

```python
#We will eliminate all unnecessary information. Personal information as well as irrelevant or duplicated
#information for our analysis.
df.drop(columns=['personName', 'category', 'organization', 'status', 'title', 'date', 'state', 'residenceStateRegion', 
                'birthMonth', 'birthDay', 'cpi_change_country', 'gross_primary_education_enrollment_country',
                 'tax_revenue_country_country', 'total_tax_rate_country', 'latitude_country', 'longitude_country'], inplace=True)
```

```python
df
```

```python
#Here, we will see which columns have NaN values so we can deal with them now
df.isnull().sum()
```

```python
#Dealing with 'age' column missing values
df['age'].isnull().sum()
```

```python
#We cannot change the data type if there are null values in the column, so we will impute the average age for the 
#65 null values and then convert column data type to int64
avg_age=df['age'].mean()
df['age'].fillna(avg_age, inplace=True)
```

```python
#Retesting
df['age'].isnull().sum()
```

```python
#Now, we will deal with the birthYear missing values
df['birthYear'].isnull().sum()
```

```python
#Given that we imputed the age, that would mean the imputed birthyear would match with the imputed age, so we will
#go ahead and proceed in the same manner with this column.
avg_YOB=df['birthYear'].mean()
df['birthYear'].fillna(avg_YOB, inplace=True)
```

```python
#Retesting
df['birthYear'].isnull().sum()
```

```python
#Given that we cannot impute city and country missing values with the mean method, and there are too many for 
#manually imputing them, we will go ahead and delete both the 'city' and 'country' observations with NaN values
df.dropna(subset=['country', 'city'], inplace=True)
```

```python
df.isnull().sum()
```

```python
#Now, we will go ahead and delete the 'birthDate' column. Having the birth year will be enough
df.drop(columns=['birthDate'], inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#Deleting the NaN values in the remaining categories, specially given the fact that big percentage of these missing 
#values are within the same observations. 
df.dropna(subset=['firstName', 'cpi_country', 'gdp_country', 'gross_tertiary_education_enrollment',
                 'life_expectancy_country', 'population_country'], inplace=True)
```

### #1. Convert 'finalWorth' to an easily understood value, and rename to 'Net Worth'¶

```python
df.rename(columns={'finalWorth': 'Net Worth ($B)'}, inplace=True)
```

```python
#Converting the number format in the Net Worth column for easier comprehension
df['Net Worth ($B)'] = df['Net Worth ($B)'].map('{:.2f}'.format).astype(float)
```

```python
df['Net Worth ($B)']
```

```python
#Dividing by 1000 so number makes sense (i.e 211,000 million equals 211 billion)
df['Net Worth ($B)'] = (df['Net Worth ($B)'] / 1000).round(1)
```

### #4. Change 'cpi_country' column name to 'Residence Country CPI Inflation Percentage' and make it look like a percentage

```python
#Renaming first
df.rename(columns={'cpi_country':'Residence Country CPI Inflation Percentage'}, inplace=True)
```

```python
df.columns
```

```python
#Subtracting by 100 so actual CPI number reads as a percentage
df['Residence Country CPI Inflation Percentage'] -= 100
```

```python
df['Residence Country CPI Inflation Percentage']=df['Residence Country CPI Inflation Percentage'].astype(float)
df['Residence Country CPI Inflation Percentage']
```

```python
#Then, we add a % sign to the end of each value through the use of a lambda function
df['Residence Country CPI Inflation Percentage'] = df['Residence Country CPI Inflation Percentage'].apply(lambda x: f'{x:.1f}%')
```

```python
df['Residence Country CPI Inflation Percentage']
```

### #5. Remove dollar sign from 'gdp_country' column, change data type from object to int64 and rename to 'Home Country GDP'

```python
##Here, we will change the the 'gdp_country' column by eliminating the '$' and then changing the data type to int64
df['gdp_country']=df['gdp_country'].str.replace('$', '')
```

```python
#First, here we remove commas and spaces from the 'gdp_country' column
df['gdp_country'] = df['gdp_country'].str.replace(',', '').str.strip()

#After that, we fill missing values with a placeholer ('0') since we cant convert NaN values to integers
df['gdp_country'] = df['gdp_country'].fillna('0')

#Finally, we convert the values to integers
df['gdp_country'] = df['gdp_country'].astype(int)
```

```python
df.rename(columns={'gdp_country':'Residence Country GDP'}, inplace=True)
```
